```python
import random
```

```python
# Define beach name sequence
beach_names = ["Beach" + str(i + 1) for i in range(2000)]
```

```python
# Define beach regions
regions = [
    "California",
    "Florida",
    "Hawaii",
    "Caribbean",
    "Australia",
    "French Polynesia",
    "Maldives",
    "Seychelles",
    "Greece",
    "Italy",
    "Thailand",
    "Indonesia",
    "Philippines",
    "Dominican Republic",
    "Cuba",
    "Mexico",
    "South Africa",
    "Sri Lanka",
    "Goa",
    "Lakshadweep",
]
```

```python
def generate_beach_data():
  beach_name = beach_names.pop(0)  # Use pop(0) to get the first element
  region = random.choice(regions)
  latitude = random.uniform(20, 60)  # Adjust range based on desired locations
  longitude = random.uniform(-180, 180)  # Adjust range based on desired locations
  sand_type = random.choice(["Fine sand", "Coarse sand", "Pebbles"])
  water_clarity = random.choice(["Crystal clear", "Clear", "Murky"])
  wave_height = random.choice(["Calm", "Small waves", "Moderate waves", "Large waves"])
  water_temp = random.uniform(20, 30)  # Adjust range based on desired locations
  lifeguard_presence = random.randint(0, 1)

  amenities = {key for key, value in [("Public Restrooms", random.randint(0, 1)),
                                            ("Showers", random.randint(0, 1)),
                                            ("Parking", random.randint(0, 1)),
                                            ("Concessions", random.randint(0, 1)),
                                            ("Rental Equipment", random.randint(0, 1))]
                 if value == 1}  # Include amenity only if value is 1

  activities = {key for key, value in [("Swimming", random.randint(0, 1)),
                                               ("Surfing", random.randint(0, 1) if wave_height != "Calm" else 0),
                                               ("Snorkeling/Diving", random.randint(0, 1) if water_clarity in ["Crystal clear", "Clear"] else 0),
                                               ("Fishing", random.randint(0, 1)),
                                               ("Sunbathing", random.randint(0, 1))]
                if value == 1}  # Include activity only if value is 1

  crowds = random.randint(1, 5)
  accessibility = random.randint(0, 1)
  popularity = random.randint(1, 5)
  cost = random.randint(50, 500)  # Adjust range based on desired cost scale

  return {
      "Beach Name": beach_name,
      "Region": region,
      "Latitude": latitude,
      "Longitude": longitude,
      "Sand Type": sand_type,
      "Water Clarity": water_clarity,
      "Wave Height": wave_height,
      "Water Temperature": water_temp,
      "Lifeguard Presence": lifeguard_presence,
      "Amenities": amenities,  # Filtered amenities (only keys)
      "Activities": activities,  # Filtered activities (only keys)
      "Crowds": crowds,
      "Accessibility": accessibility,
      "Popularity": popularity,
      "Cost": cost
  }
```

```python
# Generate 2000 beach entries
data = []
for _ in range(2000):
  data.append(generate_beach_data())
```

```python
# Print or save the data (modify as needed)
print(data)  # This will print the entire dataset
```

```python
# You can save the data to a CSV file using libraries like pandas
import pandas as pd
df = pd.DataFrame(data)
df.to_csv("beach_dataset.csv", index=False)
```
